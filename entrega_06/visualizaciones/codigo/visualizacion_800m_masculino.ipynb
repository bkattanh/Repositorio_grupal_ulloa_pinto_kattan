```python
import pandas as pd
import altair as alt

ruta_archivo = '/content/drive/MyDrive/base de datos general visualizacion.csv'
df = pd.read_csv(ruta_archivo, delimiter=';')

df_masc_800 = df[
    (df['Prueba'] == '800m') &
    (df['Categoria'].str.contains('Masculino', na=False))
].copy()

df_masc_800['Marca'] = df_masc_800['Marca'].astype(str).str.replace(',', '.').astype(float)


df_masc_800['Tipo Categoria'] = df_masc_800['Categoria'].apply(
    lambda x: 'Adulto' if 'Adulto' in x else 'Juvenil'
)

grafico_masc = alt.Chart(df_masc_800).mark_circle(size=80).encode(
    x=alt.X('Año:O', title='Año'),
    y=alt.Y('Marca:Q', title='Tiempo (segundos)', scale=alt.Scale(zero=False)),
    color=alt.Color('Tipo Categoria:N', title='Categoría', legend=alt.Legend(orient='right')),
    tooltip=['Nombre Atleta', 'Marca', 'Año', 'Competencia', 'Categoria']
).properties(
    title='Distribución de Marcas Masculinas en 800m (Adultos vs Juveniles)',
    width=600,
    height=400
).interactive()

grafico_masc
```

```python
from google.colab import files

grafico_masc.save('visualizacion_800m_masculino.html')

files.download('visualizacion_800m_masculino.html')
```
